```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/matotomoya/Downloads/gdm_materials_discovery/gnome_data/stable_materials_summary.csv")
```

```python
# Filtering the dataset for materials that contain both 'Na' and 'Si'
na_si_materials = df[df['Elements'].str.contains('Na') & df['Elements'].str.contains('Si')]
```

```python
na_si_materials
```

```python
na_si_materials.to_csv('./output/na_si_materials.csv', index=False)
```

```python
import shutil
import os

# 元のディレクトリと新しいディレクトリのパスを設定
original_directory = '/Users/matotomoya/Downloads/gdm_materials_discovery/gnome_data/by_id/'
new_directory = '/Users/matotomoya/Downloads/gdm_materials_discovery/gnome_data/na_si_cif_files/'

# 新しいディレクトリを作成（存在しない場合）
if not os.path.exists(new_directory):
    os.makedirs(new_directory)

# フィルタリングされた MaterialId のリストを取得
material_ids = na_si_materials['MaterialId'].tolist()

# 各 MaterialId に対応する .cif ファイルを検索し、新しいディレクトリにコピー
for material_id in material_ids:
    cif_file = f"{material_id}.CIF"
    original_file_path = os.path.join(original_directory, cif_file)
    new_file_path = os.path.join(new_directory, cif_file)

    # ファイルが存在する場合にのみコピー
    if os.path.exists(original_file_path):
        shutil.copy2(original_file_path, new_file_path)
```
